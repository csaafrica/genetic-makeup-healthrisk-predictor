```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import sys
```

```python
bio_data = pd.read_csv(r'C:\Users\ryanj\Desktop\CSA2025\data\anthropometric_trait_gwas.csv')
```

```python
# histogram showing the distribution of LDL_cholesterol levels, HDL_cholesterol, systolic_BP, diastolic_BP

fig, ax = plt.subplots(2,2, figsize=(15, 12))

# distribution of LDL_cholesterol levels
sns.histplot(bio_data['LDL_cholesterol'], bins=30, kde=True, ax=ax[0,0])
ax[0,0].set_title('Distribution of LDL Cholesterol Levels', fontsize=14, fontweight='bold')
ax[0,0].set_xlabel('LDL Cholesterol Level (mg/dL)')
ax[0,0].set_ylabel('Frequency')

# distribution of HDL_cholesterol levels
sns.histplot(bio_data['HDL_cholesterol'], bins=30, kde=True, ax=ax[0,1])
ax[0,1].set_title('Distribution of HDL_cholesterol', fontsize=14, fontweight='bold')
ax[0,1].set_xlabel('HDL Cholesterol Level (mg/dL)')
ax[0,1].set_ylabel('Frequency')

# distribution of systolic_BP levels
sns.histplot(bio_data['systolic_BP'], bins=30, kde=True, ax=ax[1,0])
ax[1,0].set_title('Distribution of systolic_BP', fontsize=14, fontweight='bold')
ax[1,0].set_xlabel('Systolic Blood Pressure (mmHg)')
ax[1,0].set_ylabel('Frequency')

# distribution of diastolic_BP
sns.histplot(bio_data['diastolic_BP'], bins=30, kde=True, ax=ax[1,1])
ax[1,1].set_title('Distribution of diastolic_BP', fontsize=14, fontweight='bold')
ax[1,1].set_xlabel('Diastolic Blood Pressure (mmHg)')
ax[1,1].set_ylabel('Frequency')
```

```python
bio_data.describe()
```

```python
# A relplot showing the relationship between LDL_cholesterol and BMI

sns.set_theme()

sns.relplot(
    data=bio_data,
    x="BMI", y="LDL_cholesterol",
    hue="sex", sizes=(100, 350),
);
```

```python
# scatter plot showing the relationship between diastolic blood pressure and age

sns.set_theme(style="whitegrid")

sns.relplot(
    data=bio_data,
    x="age", y="diastolic_BP",
    hue="sex", sizes=(100, 350),
);
```

```python
fig, ax = plt.subplots(1,2, figsize=(12, 5))

sns.lineplot(data=bio_data, x="age", y="diastolic_BP", hue="sex", size=None, style=None, sizes=(100, 350), ax=ax[0]);
ax[0].set_title('Diastolic Blood Pressure vs Age')

sns.lineplot(data=bio_data, x="age", y="systolic_BP", hue="sex", size=None, style=None, sizes=(100, 350), ax=ax[1]);
ax[1].set_title('Systolic Blood Pressure vs Age')


```

```python
# how BMI relates to blood pressure
fig, ax = plt.subplots(1, 2, figsize=(10, 6))

sns.scatterplot(
    data=bio_data,
    x="BMI", y="diastolic_BP",
    hue="sex", ax=ax[0]
)
ax[0].set_title("Diastolic BP against BMI")

sns.scatterplot(
    data=bio_data,
    x="BMI", y="systolic_BP",
    hue="sex", ax=ax[1]
)
ax[1].set_title("systolic BP against BMI")

plt.tight_layout() 
```

```python
# boxplot group by sex for BMI, LDL/HDL cholesterol, Blood pressure

fig, ax = plt.subplots(2, 2, figsize=(15, 10))

# Boxplot for BMI
#palette = {'Male': 'skyblue', 'Female': 'lightpink'}

sns.boxplot(x="sex", y="BMI", hue="sex", data=bio_data, ax=ax[0,0]);
ax[0,0].set_title("BMI", fontsize=14, fontweight='bold', color='blue');

# Boxplot for systolic BP
sns.boxplot(x="sex", y="systolic_BP", hue="sex", data=bio_data, ax=ax[0,1]);
ax[0,1].set_title("systolic BP", fontsize=14, fontweight='bold', color='blue');

# Boxplot for LDL cholesterol
sns.boxplot(x="sex", y="LDL_cholesterol", hue="sex", data=bio_data, ax=ax[1,0]);
ax[1,0].set_title("LDL cholesterol", fontsize=14, fontweight='bold', color='blue');

# Boxplot for HDL cholesterol
sns.boxplot(x="sex", y="HDL_cholesterol", hue="sex", data=bio_data, ax=ax[1,1]);
ax[1,1].set_title("HDL cholesterol", fontsize=14, fontweight='bold', color='blue');

# Boxplot for weight
#ax[0,1] = sns.boxplot(x="sex", y="weight", hue="sex", data=bio_data);

```

```python
selected_columns = ['LDL_cholesterol', 'HDL_cholesterol', 'systolic_BP', 'diastolic_BP', 'BMI', 'age']

snp_cols = [col for col in bio_data.columns if col.startswith('SNP_')]
selected_snps = snp_cols[:6]
corr_columns = selected_columns + selected_snps

correlation_matrix = bio_data[corr_columns].corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
```
